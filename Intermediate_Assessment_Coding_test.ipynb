```python
#Write a python program to check if the entered string is palindrome or not.
#Get the input string from user (A palindrome is a word, phrase, number, or sequence of characters that reads the same backward as forward)
#Sample input : malayalam Sample output : palindrome Sample input : computer Sample output : not palindrome


word = input("Enter a Word: ")
if word == word[::-1]:
    print(f"{word}  is palindrome")
else:
    print(f"{word}  is not palindrome")
```

```python
#Write a python program to check if the given number is divisible by 9

num = int(input("Enter a number: "))
if num % 9 == 0:
    print("The number is divisible by 9")
else:
    print("The number is not divisible by 9")
```

```python
#Write a python program to print odd numbers between 300 and 500

for i in range(300, 501):
    if i % 2 != 0:
        print(i)
```
